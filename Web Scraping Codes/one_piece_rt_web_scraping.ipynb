```python
import pandas as pd
import numpy as np
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
import time
```

## Webscrape Rotten Tomatoes Reviews Using Selenium

```python
# First, open general audience reviews on Rotten Tomatoes

url = 'https://www.rottentomatoes.com/tv/one_piece_2023/s01/reviews/all-audience'

option = Options()
option.add_argument("start-maximized")

driver = webdriver.Chrome(options=option)
driver.get(url)

# Pass Policy Agreement by clicking 'Continue'

rt_continue = driver.find_element(By.CLASS_NAME, 'ot-button-order-2')

while True:
    try:
        rt_continue.click()
    except Exception as e:
        break
```

```python
# Before webscraping reviews, click all the 'load more' buttons

load_more_2 = driver.find_element(By.CLASS_NAME, 'button-wrap')

while True:
    try:
        load_more_2.click()
        time.sleep(0.5)
    except Exception as e:
        break
```

```python
# User ID, Review Date, Info

user_ids = []
timestamps = []
reviews = []

for i in range(1, 1231):
    user_id = driver.find_element(By.XPATH, '//*[@id="main-page-content"]/div/section/div/div[2]/div[1]/review-card[{}]/rt-link[2]'.format(1*i)).text
    timestamp = driver.find_element(By.XPATH, '//*[@id="main-page-content"]/div/section/div/div[2]/div[1]/review-card[{}]/span[@slot="timestamp"]'.format(1*i)).text
    review = driver.find_element(By.XPATH, '//*[@id="main-page-content"]/div/section/div/div[2]/div[1]/review-card[{}]/drawer-more/span'.format(1*i)).text

    user_ids.append(user_id)
    timestamps.append(timestamp)
    reviews.append(review)
```

```python
# Check to make sure all of them are included

print(len(user_ids), len(timestamps), len(reviews))
```

## Create Dataframe

```python
# Create dateframe

one_piece_raw = {'User_IDs': user_ids, 
                 'Timestamps': timestamps,
                 'Reviews': reviews}

one_piece_RT_df = pd.DataFrame(one_piece_raw)

one_piece_RT_df["Source"] = 'Rotten Tomatoes'
```

```python
# Save it to csv

one_piece_RT_df.to_csv('one_piece_RT.csv')
```

```python
# End Selenium

driver.close()
driver.quit()
```
